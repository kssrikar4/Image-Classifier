```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
import numpy as np
import os

# Configuration
IMAGE_SIZE = (160, 160)
BATCH_SIZE = 32
SHUFFLE_BUFFER_SIZE = 1000
AUTOTUNE = tf.data.AUTOTUNE
EPOCHS_FINE_TUNE = 40
BASE_MODEL_TRAINABLE_LAYERS = 70
MODEL_SAVE_PATH = './stanford_dogs_model'
PLOTS_SAVE_DIR = './training_plots_EfficientNetB0'
DROPOUT_RATE = 0.3

os.makedirs(MODEL_SAVE_PATH, exist_ok=True)
os.makedirs(PLOTS_SAVE_DIR, exist_ok=True)

# 1. Load and Prepare Data
print("Loading Stanford Dogs dataset...")
(ds_train, ds_test), ds_info = tfds.load(
    'stanford_dogs',
    split=['train', 'test'],
    shuffle_files=True,
    with_info=True,
    as_supervised=True,
)

num_classes = ds_info.features['label'].num_classes
print(f"Number of classes: {num_classes}")

data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(factor=0.1),
    tf.keras.layers.RandomZoom(height_factor=0.2, width_factor=0.2),
])

def resize_image(image, label):
    image = tf.image.resize(image, IMAGE_SIZE)
    return image, label

def apply_augmentation(image, label):
    image = data_augmentation(image, training=True)
    image = tf.image.random_brightness(image, max_delta=50)
    image = tf.image.random_contrast(image, lower=0.7, upper=1.3)
    image = tf.image.random_saturation(image, lower=0.7, upper=1.3)
    image = tf.image.random_hue(image, max_delta=0.1)
    image = tf.clip_by_value(image, 0.0, 255.0)
    return image, label

def efficientnet_preprocess(image, label):
    image = tf.keras.applications.efficientnet.preprocess_input(image)
    return image, label

ds_train = ds_train.map(resize_image, num_parallel_calls=AUTOTUNE)
ds_train = ds_train.map(apply_augmentation, num_parallel_calls=AUTOTUNE)
ds_train = ds_train.map(efficientnet_preprocess, num_parallel_calls=AUTOTUNE)
ds_train = ds_train.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE).prefetch(AUTOTUNE)

ds_test = ds_test.map(resize_image, num_parallel_calls=AUTOTUNE)
ds_test = ds_test.map(efficientnet_preprocess, num_parallel_calls=AUTOTUNE)
ds_test = ds_test.batch(BATCH_SIZE).prefetch(AUTOTUNE)

print("Dataset preprocessing complete with fixed TF-native augmentation.")

# 2. Choose and Build the Model
print("Building the model using EfficientNetB0 with Dropout...")

base_model = tf.keras.applications.EfficientNetB0(
    input_shape=(IMAGE_SIZE[0], IMAGE_SIZE[0], 3),
    include_top=False,
    weights='imagenet'
)
base_model.trainable = False

global_average_layer = tf.keras.layers.GlobalAveragePooling2D()
dropout_layer = tf.keras.layers.Dropout(DROPOUT_RATE)
prediction_layer = tf.keras.layers.Dense(num_classes, activation='softmax')

model = tf.keras.Sequential([
    base_model,
    global_average_layer,
    dropout_layer,
    prediction_layer
])

# 3. Set up the Training Process
print("Compiling the model for initial training...")
base_learning_rate = 0.0001
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=base_learning_rate),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              metrics=['accuracy'])

model.summary()

print("Starting initial training (training only the top layers)...")
initial_epochs = 10
history = model.fit(ds_train,
                    epochs=initial_epochs,
                    validation_data=ds_test)

print("Initial training complete.")

# 4. Fine-tuning
print("Starting fine-tuning (unfreezing and training some base model layers)...")

base_model.trainable = True
for layer in base_model.layers[:-BASE_MODEL_TRAINABLE_LAYERS]:
    layer.trainable = False

fine_tune_learning_rate = base_learning_rate / 10
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
              optimizer=tf.keras.optimizers.Adam(learning_rate=fine_tune_learning_rate),
              metrics=['accuracy'])

model.summary()

fine_tune_callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=15, monitor='val_loss', restore_best_weights=True),
    tf.keras.callbacks.ModelCheckpoint(filepath=os.path.join(MODEL_SAVE_PATH, 'EfficientNetB0_best_model.keras'),
                                       monitor='val_accuracy',
                                       save_best_only=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=7, verbose=1, min_lr=fine_tune_learning_rate * 0.01)
]

total_epochs = initial_epochs + EPOCHS_FINE_TUNE
history_fine_tune = model.fit(ds_train,
                                epochs=total_epochs,
                                initial_epoch=initial_epochs,
                                validation_data=ds_test,
                                callbacks=fine_tune_callbacks)

print("Fine-tuning complete.")

# 5. Save the Model
print(f"Saving the final trained model state to {os.path.join(MODEL_SAVE_PATH, 'EfficientNetB0_final_model.keras')}...")
model.save(os.path.join(MODEL_SAVE_PATH, 'EfficientNetB0_final_model.keras'))
print("Final model state saved.")
print(f"Best model saved during training can be found at {os.path.join(MODEL_SAVE_PATH, 'EfficientNetB0_best_model.keras')}")

# 6. Visualize Training
print(f"Generating training plots and saving to {PLOTS_SAVE_DIR}...")

acc = history.history['accuracy'] + history_fine_tune.history['accuracy']
val_acc = history.history['val_accuracy'] + history_fine_tune.history['val_accuracy']
loss = history.history['loss'] + history_fine_tune.history['loss']
val_loss = history.history['val_loss'] + history_fine_tune.history['val_loss']

epochs_range = range(len(acc))

plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.plot([initial_epochs-1, initial_epochs-1], plt.ylim(), label='Start Fine Tuning', linestyle='--')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.plot([initial_epochs-1, initial_epochs-1], plt.ylim(), label='Start Fine Tuning', linestyle='--')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

plt.savefig(os.path.join(PLOTS_SAVE_DIR, 'training_validation_curves.png'))
plt.show()

print("Training plots saved.")
```

```python
import os
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

model = tf.keras.models.load_model('./stanford_dogs_model/EfficientNetB0_final_model.keras')

(ds_train, ds_test), ds_info = tfds.load(
    'stanford_dogs', 
    split=['train', 'test'],
    with_info=True,
    as_supervised=True
)

ds_all = ds_train.concatenate(ds_test)

IMG_SIZE = model.input_shape[1:3]

def preprocess(image, label):
    image = tf.image.resize(image, IMG_SIZE)
    image = tf.keras.applications.efficientnet.preprocess_input(image)
    return image, label

autotune = tf.data.AUTOTUNE
batch_size = 32

ds_all = ds_all.map(preprocess, num_parallel_calls=autotune)
ds_all = ds_all.batch(batch_size).prefetch(autotune)

pred_probs = model.predict(ds_all, verbose=1)
pred_labels = np.argmax(pred_probs, axis=1)

true_labels = np.concatenate([y.numpy() for _, y in ds_all], axis=0)

class_names = ds_info.features['label'].names

print(classification_report(true_labels, pred_labels, target_names=class_names))

cm = confusion_matrix(true_labels, pred_labels)
cm_norm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  
```
